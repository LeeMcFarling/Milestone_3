# 2. Weekly graph question

Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?


#### **Answer:**
- **Histogram**
    - Pros: The strength of the histogram is that it allows us to see *how* the data is distributed. For example, is the distribution a normal distribution or is it skewed either left or right (what direction are the tails?). Additionally, the histogram allows us to visually see the variance of the dataset. In this case, it would allow us to ascertain whether the drug prices are clustered within a narrow range or whether they are spread out, etc. 
    
  - Cons: While a histogram provides a general sense of distribution, it does not explicitly highlight outliers. It also lacks precise indicators of central tendency, such as the exact mean or median. While one can estimate these values by observing the distribution (e.g., estimating that the mean price is around 10), the histogram does not provide a definitive reference point. Moreover, the choice of bin width can impact how the data appears, potentially obscuring or exaggerating trends.
- **Box and Whisker Plot**
    - Pros: A boxplot explicitly identifies outliers and defines the thresholds beyond which data points are considered extreme. It also clearly displays key summary statistics, including quartiles and the median, making it easier to compare price distributions at a glance. This is especially useful for understanding the range and skewness of drug prices. (At least it gives us a line to visually represent it whereas we kind of need to guestimate using the histogram)

    - Cons: While a boxplot provides a concise summary of the dataset’s distribution, it does not reveal finer details about how data points are distributed within each quartile. Unlike a histogram, it does not show whether the data follows a smooth curve, has gaps, or clusters at specific price points. Additionally, if there are many outliers, the plot can become cluttered, making interpretation more difficult

- **Which is better?**
  -  As for which plot type is better, it would depend on what information you are trying to extract from the dataset. If you are interested in getting a quick summary for how the data is distributed, then a histogram would be a better plot to use. If you wanted to get a handle on what the outliers are in the dataset, and where those specific boundaries might be, then a box and whisker plot would be a better choice for that. 

- **How Would I Improve it / them?**
- I would improve the histogram by adjusting the bin edges so they align precisely with the axis ticks. This adjustment would make it easier to interpret the distribution by clearly indicating where each bin starts and ends, enhancing readability and accuracy in identifying data groupings.

- Additionally, for the box and whisker plot, I would add labels for the quartiles and min/max. Although we can see the lines for these measures now, adding these labels would help us to fully quantify where these measures are in the data. 

- For *both* plots, I would clearly label the axises on both the x and y axis in order for users to be able to clearly identify what they are looking at in terms of drug prices for this pharmacy and how these drug prices are distributed. Additionally, I would add titles to each as well.

### **Visualizations**

First, I want to get an idea of *how many* injuries occured of each type: 

```python

injury_counts = InjuryRecord.groupby(['InjuryLength', 'BodyPart']).size().reset_index(name='Injury Count')
injury_lengths = ["1+", "7+", "28+", "42+"]
injury_labels = ["1-Day +", "1-Week +", "1-Month +", "1.5 Months +"]


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))  # 2x2 grid
axes = axes.flatten()


for i, (injury_key, label) in enumerate(zip(injury_lengths, injury_labels)):
    ax = axes[i]
    subset = injury_counts[injury_counts['InjuryLength'] == injury_key]
    
    if not subset.empty:
        ax.bar(subset['BodyPart'], subset['Injury Count'], color='steelblue')
        ax.set_title(f"{label} Injuries")
        ax.set_xlabel("Body Part")
        ax.set_ylabel("Injury Count")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
def chart_totals(df, column):
    plt.figure(figsize=(12, 4))
    df[column].value_counts().plot(kind='bar')

    plt.title(f"Count of Each Category in {column}")
    plt.xlabel(column)
    plt.ylabel("Total Count")
    plt.show()
```
